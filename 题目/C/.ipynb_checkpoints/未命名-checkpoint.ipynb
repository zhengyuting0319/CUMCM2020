```python
import torch
from torch import nn

print(torch.__version__)
```

```python
class MLP(nn.Module):
    def __init__(self, **kwargs):
        super(MLP, self).__init__(**kwargs)
        self.net = nn.Sequential(nn.Linear(8, 30)
                                ,nn.ReLU()
                                ,nn.Linear(30, 4)
                                ,nn.ReLU()) 

    def forward(self, x):
        return self.net(x)

net = MLP()

print(net)
```

```python
# define our training function
def train(model, device, train_loader, optimizer, epoch):
    model.train()
    loss_fn = nn.mse() #CrossEntropyLoss()
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        output = model(data)
        loss = loss_fn(output, target)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if batch_idx % 200 == 0:
            print ('Train Epoch: {}, Batch Index: {}, Loss: {}'.format(
                epoch, batch_idx, loss.item()))
```

```python
# define out testing function
def test(model, device, test_loader):
    model.eval()
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            pred = output.argmax(dim=1,) #(bsz,)
            correct += pred.eq(target).sum().item()
    print (" Test Accuracy: {}%".format(
            correct/len(test_loader.dataset)*100))
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

BSZ = 30
LR = 1e-3
EPOCH = 1

model = net.to(device)
optimizer = optim.Adam(model.parameters(), lr=LR)
for epoch in range(1, EPOCH+1):
    train(net, device, train_loader, optimizer, epoch)
    test(net, device, test_loader)

#     # uncomment this line to save the trained model 
#     torch.save(model.state_dict(), "mnist_cnn.pt")
```
